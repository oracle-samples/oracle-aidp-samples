# Kafka Connector Samples

Read-only ingestion samples for the `KAFKA` connector.


## Ingestion Sample

Kafka crawling is ingestion-style only for these samples. Replace the bootstrap servers and topic values before running.


```python
kafka_df = spark.read.format("aidataplatform") \
    .option("type", "KAFKA") \
    .option("bootstrap.servers", "<BOOTSTRAP_SERVERS>") \
    .option("ssl.enabled", "true") \
    .option("user.name", "<USERNAME>") \
    .option("password", "<PASSWORD>") \
    .option("schema", "<TOPIC_NAME>") \
    .option("message", "<TOPIC_NAME>:<PARTITION>") \
    .option("host.name.verification", "true") \
    .load()

kafka_df.show()
```

## Connector Options

| Parameter name | Valid values | Mandatory | Description |
| --- | --- | --- | --- |
| `type` | `KAFKA` | Y | Data source type. |
| `bootstrap.servers` | `host1:port1,host2:port2,...` | Y | Comma-separated list of Kafka bootstrap servers. |
| `ssl.enabled` | `true`, `false` | N | Enable SSL or TLS for Kafka connectivity. |
| `truststore.password` | Truststore password | N | Truststore password when SSL is used. |
| `truststore.file.path` | Valid file path | N | Truststore file path when SSL is used. |
| `user.name` | Valid username | N | Username for Kafka PLAIN authentication. |
| `password` | Valid password | N | Password for Kafka PLAIN authentication. |
| `schema` | Topic name | Y | Kafka topic name when using schema-style selection. |
| `message` | Topic:partition or blank | N | Topic-partition selector. If omitted, all partitions are read. |
| `host.name.verification` | `true`, `false` | N | Enable or disable hostname verification during SSL or TLS handshake. |
| `read.as.string` | `true`, `false` | N | Read Kafka payloads as string values instead of raw binary values. |
| `starting.offset` | `earliest` or explicit offset config | N | Starting offset configuration. |
| `ending.offsets` | `latest` or explicit offset config | N | Ending offset configuration for batch reads. |
| `fetch.offset.num.retries` | Positive integer | N | Number of retries when fetching offsets. |
| `offset.retry.interval.ms` | Positive integer | N | Retry interval in milliseconds when fetching offsets. |
| `poll.timeouts.ms` | Positive integer | N | Kafka consumer poll timeout in milliseconds. |
| `fail.on.data.loss` | `true`, `false` | N | Fail the read if data loss is detected. |
