# MySQL HeatWave Connector Samples

Read-only ingestion samples for the `MYSQL_HEATWAVE` connector.


## Ingestion Sample

Replace placeholders such as `<HOST>`, `<USERNAME>`, `<PASSWORD>`, `<SCHEMA>`, and `<TABLE_NAME>` with values from your environment before running the notebook.


```python
# Read from MySQL HeatWave
heatwave_df = spark.read.format("aidataplatform") \
    .option("type", "MYSQL_HEATWAVE") \
    .option("host", "<HOST>") \
    .option("port", "<PORT>") \
    .option("user.name", "<USERNAME>") \
    .option("password", "<PASSWORD>") \
    .option("schema", "<SCHEMA>") \
    .option("table", "<TABLE_NAME>") \
    .load()

heatwave_df.show()
```

## Pushdown SQL Sample

Use `pushdown.sql` when you want the connector to execute a complete source SQL query instead of building the query from `schema` and `table` options.


```python
# Full SQL pushdown for MySQL HeatWave
heatwave_df_pushdown = spark.read.format("aidataplatform") \
    .option("type", "MYSQL_HEATWAVE") \
    .option("host", "<HOST>") \
    .option("port", "<PORT>") \
    .option("user.name", "<USERNAME>") \
    .option("password", "<PASSWORD>") \
    .option("pushdown.sql", "SELECT * FROM <SCHEMA>.<TABLE_NAME> WHERE <COLUMN_NAME> = '<VALUE>'") \
    .load()

heatwave_df_pushdown.show()
```

## Connector Options

| Parameter name | Valid values | Mandatory | Description |
| --- | --- | --- | --- |
| `type` | `MYSQL_HEATWAVE` | Y | Data source type. |
| `host` | Valid hostname or IP | Y | Database host. |
| `port` | Valid port number | Y | Database port. |
| `database.name` | Database or catalog name | N | Optional database name when needed. |
| `user.name` | Valid DB user | Y | Login user used for connectivity. |
| `password` | Valid DB password | Y | Password for `user.name`. |
| `schema` | Valid schema name | Y | Schema to read from. |
| `table` | Valid table name | Y for table access | Table name inside the schema. |
| `view` | Valid view name | N | View name inside the schema. |
| `sql` | Valid SQL query | N | Custom SQL query for reads. |
| `mysql.time.read.mode` | `string` | N | Read MySQL `TIME` values in compatibility mode when duration-like values such as `99:59:59` or `-12:30:00` should stay as strings instead of being coerced into wall-clock time. |
| `fetch.size` | Positive integer | N | Number of records fetched per round trip for reads. |
| `row.limit` | Positive integer | N | Limits the number of rows read. |
| `partition.column` | Column name | N | Column used for range-based partitioned reads. |
| `partition.num` | Integer >= 1 | N | Number of partitions to use. |
| `partition.lower` | Numeric value or ISO-8601 date/time | N | Lower bound for range partitioning. |
| `partition.upper` | Numeric value or ISO-8601 date/time | N | Upper bound for range partitioning. |
| `pushdown.columns` | Comma-separated column names | N | Restricts the read to only the requested columns. |
| `pushdown.filter` | SQL predicate | N | Predicate pushed down to the source. |
| `pushdown.sql` | SQL query | N | Full SQL pushdown query. |
| `login.timeout.seconds` | Integer from `1` to `300` | N | JDBC login timeout in seconds. |
