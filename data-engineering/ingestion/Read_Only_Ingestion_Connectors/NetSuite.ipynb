# NetSuite Connector Samples

Read-only ingestion samples for the `NETSUITE` SuiteAnalytics Connect connector. Replace placeholders before running a sample.

## Username and Password Read

Use this authentication path when SuiteAnalytics Connect is configured for a NetSuite username and password.

```python
netsuite_df = spark.read.format("aidataplatform") \
    .option("type", "NETSUITE") \
    .option("host", "<SUITEANALYTICS_CONNECT_HOST>") \
    .option("port", "<PORT>") \
    .option("ns.account.id", "<ACCOUNT_ID>") \
    .option("ns.role.id", "<ROLE_ID>") \
    .option("user.name", "<USERNAME>") \
    .option("password", "<PASSWORD>") \
    .option("schema", "<SCHEMA>") \
    .option("table", "<TABLE_NAME>") \
    .load()

netsuite_df.show()
```

## OAuth Access-Token Read

Use this authentication path when you have a fresh SuiteAnalytics Connect OAuth 2.0 access token. Do not include both `ns.access.token` and username/password in the same request.

```python
netsuite_token_df = spark.read.format("aidataplatform") \
    .option("type", "NETSUITE") \
    .option("host", "<SUITEANALYTICS_CONNECT_HOST>") \
    .option("port", "<PORT>") \
    .option("ns.account.id", "<ACCOUNT_ID>") \
    .option("ns.role.id", "<ROLE_ID>") \
    .option("ns.access.token", "<ACCESS_TOKEN>") \
    .option("schema", "<SCHEMA>") \
    .option("table", "<TABLE_NAME>") \
    .load()

netsuite_token_df.show()
```

## Pushdown SQL

Use `pushdown.sql` to run a source query through SuiteAnalytics Connect.

```python
netsuite_pushdown_df = spark.read.format("aidataplatform") \
    .option("type", "NETSUITE") \
    .option("host", "<SUITEANALYTICS_CONNECT_HOST>") \
    .option("port", "<PORT>") \
    .option("ns.account.id", "<ACCOUNT_ID>") \
    .option("ns.role.id", "<ROLE_ID>") \
    .option("ns.access.token", "<ACCESS_TOKEN>") \
    .option("pushdown.sql", "SELECT * FROM <SCHEMA>.<TABLE_NAME> WHERE <COLUMN_NAME> = '<VALUE>'") \
    .load()

netsuite_pushdown_df.show()
```

## Connector Options

| Parameter name | Valid values | Mandatory | Description |
| --- | --- | --- | --- |
| `type` | `NETSUITE` | Y | Data source type. |
| `host`, `port` | SuiteAnalytics Connect endpoint | Y | NetSuite endpoint host and port. |
| `ns.account.id`, `ns.role.id` | Valid NetSuite account and role IDs | Y | Account and role used for the connection. |
| `user.name`, `password` | Valid NetSuite credentials | Y for username/password authentication | Credentials for username/password authentication. |
| `ns.access.token` | Fresh OAuth access token | Y for access-token authentication | OAuth 2.0 access token used instead of username/password. |
| `schema`, `table`, `sql` | Valid source selection | `schema` and `table` are Y for table access | Select a table or query. |
| `fetch.size`, `row.limit` | Positive integer | N | Read batching and limiting. |
| `partition.column`, `partition.num`, `partition.lower`, `partition.upper` | Partition settings | N | Range-partitioned reads. |
| `pushdown.columns`, `pushdown.filter`, `pushdown.sql` | Valid projection, predicate, or SQL | N | Push work down to NetSuite. |
| `login.timeout.seconds` | Integer from `1` to `300` | N | JDBC login timeout. |

NetSuite is read-only in this connector; no write sample is provided.
