# REST Connector Samples

Read-only ingestion samples for the `GENERIC_REST` connector.


## Ingestion Sample

Use REST connector options to call an API described by a manifest. Replace placeholder values before running.


```python
# REST API - read using Generic REST connector
rest_df = spark.read.format("aidataplatform") \
    .option("type", "GENERIC_REST") \
    .option("base.url", "<BASE_URL>") \
    .option("manifest.url", "<MANIFEST_URL>") \
    .option("auth.type", "basic") \
    .option("user.name", "<USERNAME>") \
    .option("password", "<PASSWORD>") \
    .option("schema", "<SCHEMA_OR_MANIFEST_PATH>") \
    .option("api", "<API_OPERATION>") \
    .option("derived.property.orderNo", "<ORDER_NO>") \
    .load()

rest_df.show()
```

## Manifest Path Samples

Use `manifest.path` when the REST manifest JSON is stored in the workspace or in a mounted volume instead of being served from a URL.


### Using Workspace Manifest Path

This option reads the REST manifest from a workspace file path.


```python
workspace_manifest_df = spark.read.format("aidataplatform") \
    .option("type", "GENERIC_REST") \
    .option("base.url", "<BASE_URL>") \
    .option("manifest.path", "/Workspace/rest-manifests/orders_manifest.json") \
    .option("auth.type", "basic") \
    .option("user.name", "<USERNAME>") \
    .option("password", "<PASSWORD>") \
    .option("schema", "<SCHEMA_OR_MANIFEST_PATH>") \
    .option("api", "<API_OPERATION>") \
    .option("derived.property.orderNo", "<ORDER_NO>") \
    .load()

workspace_manifest_df.show()
```

### Using Volumes Manifest Path

This option reads the REST manifest from a mounted volume.


```python
volume_manifest_df = spark.read.format("aidataplatform") \
    .option("type", "GENERIC_REST") \
    .option("base.url", "<BASE_URL>") \
    .option("manifest.path", "/Volumes/default/connectors/rest-manifests/orders_manifest.json") \
    .option("auth.type", "basic") \
    .option("user.name", "<USERNAME>") \
    .option("password", "<PASSWORD>") \
    .option("schema", "<SCHEMA_OR_MANIFEST_PATH>") \
    .option("api", "<API_OPERATION>") \
    .option("derived.property.orderNo", "<ORDER_NO>") \
    .load()

volume_manifest_df.show()
```

## Connector Options

| Parameter name | Valid values | Mandatory | Description |
| --- | --- | --- | --- |
| `type` | `GENERIC_REST` | Y | Data source type. |
| `base.url` | Valid base URL | Y | Base URL of the target REST source. |
| `manifest.url` | Valid URL | Y, unless `manifest.path`, `manifest.file.path`, or `manifest.content` is used | Public URL that returns the manifest JSON. |
| `manifest.path` | `/Workspace/<path>` or `/Volumes/<path>` | Y, unless `manifest.url`, `manifest.file.path`, or `manifest.content` is used | Workspace or volume path used to read manifest content. |
| `manifest.file.path` | Valid file path | Y, unless `manifest.url`, `manifest.path`, or `manifest.content` is used | Local path used to read manifest content. |
| `manifest.content` | Manifest JSON | Y, unless `manifest.url`, `manifest.path`, or `manifest.file.path` is used | Inline manifest JSON content. |
| `auth.type` | `basic`, `no.auth`, `oauth2` | Y | Authentication type for the REST call. |
| `auth.header` | Header name or header value form supported by the connector | N | Optional auth header override. |
| `user.name` | Valid username | Y for `basic` | Username for basic-auth requests. |
| `password` | Valid password | Y for `basic` | Password for basic-auth requests. |
| `schema` | Manifest path | Y | Path or resource in the manifest file. |
| `api` | Valid operation name from manifest | Y | Operation name from the manifest file. |
| `fetch.size` | Positive integer | N | Number of records read in one batch. |
| `derived.property.<name>` | Valid string value | N | Derived property value for path params, query params, or headers defined in the manifest. |
| `partition.column` | Connector-supported partition dimension | N | Used for partitioned REST reads. |
| `partition.num` | Integer >= 1 | N | Number of partitions to use. |
| `partition.lower` | Numeric or string lower bound | N | Lower bound for partitioned reads. |
| `partition.upper` | Numeric or string upper bound | N | Upper bound for partitioned reads. |
| `oauth.access.token.url` | Valid token URL | Y for `oauth2` | Access-token endpoint for OAuth2. |
| `oauth.client` | OAuth client ID | Y for `oauth2` | OAuth client identifier. |
| `oauth.client.secret` | OAuth client secret | Y for `oauth2` | OAuth client secret. |
| `oauth.client.auth.type` | Connector-supported client auth type | N | OAuth client auth type. |
| `oauth.scope` | OAuth scope string | N | OAuth scope. |
| `oauth.grant` | Connector-supported grant type | N | OAuth grant type. |
| `oauth.header.prefix` | Header prefix string | N | Prefix added before the OAuth access token in the header. |
