# Snowflake Connector Samples

Read-only samples for the `SNOWFLAKE` connector. Snowflake write operations are not supported in the 4.0 release. Replace all placeholders before running a sample.


## Basic Authentication Read


```python
snowflake_df = spark.read.format("aidataplatform") \
    .option("type", "SNOWFLAKE") \
    .option("host", "<SNOWFLAKE_ACCOUNT_OR_HOST>") \
    .option("port", "443") \
    .option("database.name", "<DATABASE_NAME>") \
    .option("authentication.method", "Basic") \
    .option("user.name", "<USERNAME>") \
    .option("password", "<PASSWORD>") \
    .option("warehouse", "<WAREHOUSE>") \
    .option("role", "<ROLE>") \
    .option("schema", "<SCHEMA>") \
    .option("table", "<TABLE_NAME>") \
    .load()

snowflake_df.show()
```

## Key-Pair Authentication Read

Provide either `private.key.file` or `private.key.content`. `private.key.pass.phrase` is optional when the private key is encrypted.


```python
snowflake_keypair_df = spark.read.format("aidataplatform") \
    .option("type", "SNOWFLAKE") \
    .option("host", "<SNOWFLAKE_ACCOUNT_OR_HOST>") \
    .option("port", "443") \
    .option("database.name", "<DATABASE_NAME>") \
    .option("authentication.method", "KeyPair") \
    .option("user.name", "<USERNAME>") \
    .option("warehouse", "<WAREHOUSE>") \
    .option("role", "<ROLE>") \
    .option("private.key.file", "<PRIVATE_KEY_FILE_PATH>") \
    .option("private.key.pass.phrase", "<PRIVATE_KEY_PASSPHRASE>") \
    .option("schema", "<SCHEMA>") \
    .option("table", "<TABLE_NAME>") \
    .load()

snowflake_keypair_df.show()

# For an inline private key, replace private.key.file with:
# .option("private.key.content", "<PRIVATE_KEY_CONTENT>")
```

## External Catalog and `catalog.id` Read

After creating a Snowflake external catalog in **Master Catalogs**, read through a three-part name or reuse the connection with `catalog.id`. These are read-only examples.


```python
catalog_snowflake_df = spark.table("<CATALOG_NAME>.<SCHEMA>.<TABLE_NAME>")
catalog_snowflake_df.show()

catalog_id_snowflake_df = spark.read.format("aidataplatform") \
    .option("catalog.id", "<CATALOG_ID>") \
    .option("schema", "<SCHEMA>") \
    .option("table", "<TABLE_NAME>") \
    .load()

catalog_id_snowflake_df.show()
```

## Pushdown SQL


```python
snowflake_pushdown_df = spark.read.format("aidataplatform") \
    .option("type", "SNOWFLAKE") \
    .option("host", "<SNOWFLAKE_ACCOUNT_OR_HOST>") \
    .option("port", "443") \
    .option("database.name", "<DATABASE_NAME>") \
    .option("authentication.method", "Basic") \
    .option("user.name", "<USERNAME>") \
    .option("password", "<PASSWORD>") \
    .option("warehouse", "<WAREHOUSE>") \
    .option("pushdown.sql", "SELECT * FROM <SCHEMA>.<TABLE_NAME> LIMIT 10") \
    .load()

snowflake_pushdown_df.show()
```

## Connector Options

| Parameter name | Valid values | Mandatory | Description |
| --- | --- | --- | --- |
| `type` | `SNOWFLAKE` | Y for inline configuration | Data source type. |
| `catalog.id` | Existing external catalog identifier | N | Reuses connection details from an external catalog. |
| `host`, `port`, `database.name` | Snowflake account/host, port, database | Y for inline configuration | Snowflake endpoint and database. |
| `authentication.method` | `Basic`, `KeyPair` | Y | Authentication mode. |
| `user.name` | Valid Snowflake user | Y | Snowflake login user. |
| `password` | Valid password | Y for `Basic` | Password for basic authentication. |
| `private.key.file`, `private.key.content` | Private key path or content | One is Y for `KeyPair` | Private key for key-pair authentication. |
| `private.key.pass.phrase` | Valid passphrase | N | Passphrase for an encrypted private key. |
| `warehouse` | Valid warehouse | Y | Warehouse used for the session. |
| `role` | Valid role | N | Snowflake role for the session. |
| `schema`, `table`, `view`, `sql` | Valid source selection | `schema` and `table` are Y for table access | Select a table, view, or query. |
| `fetch.size`, `row.limit` | Positive integer | N | Read batching and limiting. |
| `partition.column`, `partition.num`, `partition.lower`, `partition.upper` | Partition settings | N | Range-partitioned reads. |
| `pushdown.columns`, `pushdown.filter`, `pushdown.sql` | Valid projection, predicate, or SQL | N | Push work down to Snowflake. |
| `login.timeout.seconds` | Integer from `1` to `300` | N | JDBC login timeout. |
